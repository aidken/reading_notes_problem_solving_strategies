```python
import random

x_start = int( random.random() *100 )
y_start = int( random.random() *100 )

if y_start > x_start:
    temp = y_start
    y_start = x_start
    x_start = temp

f'{y_start=}, {x_start=}'
```

```python
def generate_next_x(x, y) -> float:
    return ( x + y ) / 2

def generate_next_y(x, y) -> float:
    return ( 2 * x * y ) / ( x + y )

x = [x_start]
y = [y_start]

for i in range(10):
    x.append( generate_next_x(x[-1], y[-1]) )
    y.append( generate_next_y(x[-1], y[-1]) )
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()        # Create a figure containing a single Axes.
ax.plot(x, y)                   # Plot some data on the Axes.
# plt.show()
```
